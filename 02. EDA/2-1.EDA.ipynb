### Plotting with categorical variables

```python
import matplotlib.pyplot as plt

class_labels = data[2]
values = data[3]

plt.title('title')
plt.xlabel('x')
plt.ylabel('y')
plt.figure(figsize=(60, 10))
plt.subplot(131)
plt.xticks(rotation=45)
plt.yticks()
plt.bar(class_labels, values)

# https://kite.com/python/answers/how-to-rotate-axis-labels-in-matplotlib-in-python
```

```python
# 참고 코드
# https://stackoverflow.com/questions/30228069/how-to-display-the-value-of-the-bar-on-each-bar-with-pyplot-barh

import os
import numpy as np
import matplotlib.pyplot as plt

class_labels = data[2]
values = data[3]

fig, ax = plt.subplots()
width = 0.75  # the width of the bars
ind = np.arange(len(values)) # the xlocation for the groups
ax.barh(ind, values, width, color="blue")
ax.set_yticks(ind+width/2)
ax.set_yticklabels(class_labels, minor=False)
plt.title('Num of Smaples per clas')
plt.xlabel('Number')
plt.ylabel('Class')
for i, v in enumerate(values):
    ax.text(v + 3, i + .25, str(v), color='blue', fontweight='bold')

plt.savefig(os.path.join('Malimg_Samples.png'), dpi=300, format='png', bbox_inches='tight')
```
